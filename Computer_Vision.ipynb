# Cropping Image

```python
image = cv2.imread("sq.png")
```

```python
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
start_row, end_row = 50, 200
```

```python
start_col, end_col = 200, 300
```

```python
cropped = image_rgb[start_row : end_row, start_col : end_col]
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10,4))

axes[0].imshow(image_rgb)
axes[0].set_title("Original Image")
axes[0].axis("off")

axes[1].imshow(cropped)
axes[1].set_title("Cropped Image")
axes[1].axis("off")

plt.show()
```

# Resize Image

```python
image = cv2.imread("sq.png")
```

```python
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
height, width = image.shape[:2]
```

```python
center = (width//2, height//2)
```

```python
rot_matrix = cv2.getRotationMatrix2D(center, 45, 1)
```

```python
rotated = cv2.warpAffine(image_rgb, rot_matrix, (width, height))
```

```python
rotated_rgb = cv2.cvtColor(rotated, cv2.COLOR_BGR2RGB)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10,4))

axes[0].imshow(image_rgb)
axes[0].set_title("Original Image")
axes[0].axis("off")

axes[1].imshow(rotated)
axes[1].set_title("Rotated Image")
axes[1].axis("off")

plt.show()
```

```python
# Create white canvas
flag = np.ones((300, 450, 3), dtype=np.uint8) * 255

# Saffron stripe
cv2.rectangle(flag, (0, 0), (450, 100), (0, 128, 255), -1)

# White stripe (already white)

# Green stripe
cv2.rectangle(flag, (0, 200), (450, 300), (0, 128, 0), -1)

# Ashoka Chakra
center = (225, 150)
radius = 35

cv2.circle(flag, center, radius, (255, 0, 0), 2)

# Draw 24 spokes
for i in range(24):
    angle = i * 15 * np.pi / 180
    x = int(center[0] + radius * np.cos(angle))
    y = int(center[1] + radius * np.sin(angle))
    cv2.line(flag, center, (x, y), (255, 0, 0), 1)

# Display flag
cv2.imshow("Indian Flag", flag)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
